<h><b>Step 1: Loading libraries and dataset</b></h>

```python
#Loading matplot,seaborne,numpy pandas and dataset

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.datasets import load_iris
```

```python
#loading the iris dataset and its features into the memory
iris = load_iris()
```

```python
#creating a dataframe for iris
df = pd.DataFrame(iris.data, columns = iris.feature_names)
df['target'] = iris.target
print("Iris DataSet:")
df
```

<h><b>Step 3: Making new Features and decoding relationship among variables/values</b></h>
<p>Now using different visualization techniques to find if there is any correlation among existing features, and outliers in the data, and if new features can be made or not</p>
<p>-----------------------------------------------------------</p>

<h>Using Heatmap for finding correlation:</h>



```python
#finding correlation first for the features 

correlation = df.drop(columns="target").corr()

#vizualizing the correlation 

sns.heatmap(correlation,annot=True,cmap="coolwarm")
plt.title("Analysis of Features Relationship",fontsize=16, weight='bold')
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
```

<p>Heatmap shows:</p>
<li>Petal length is highly correlated with petal width and sepal length</li>
<li>Sepal lenght is highly correlated with petal width</li>
<li>The rest are negatively correlated with each other</li>
<p><b>Conclusion: </b>By using Petal length and Petal width, Petal area can be determined</p>



```python
#making new features
df["flower area"] = df["petal length (cm)"] * df["petal width (cm)"]
df.head(36)
```

```python
print("Flower with maximum area:")
df.max()
```

```python
print("Flower with minimum area:")
df.min()
```

Making Scatter Plot:

```python
scat = df.drop(columns="target")
sns.scatterplot(data=scat,x="petal length (cm)",y="petal width (cm)",hue="flower area",legend = True,s=60)
plt.title("Relationship between length and width of a petal",fontsize=16, weight='bold')
plt.xlabel("Petal length")
plt.ylabel("Petal width")
plt.grid(True)
plt.show()
```

<b>Conclusion Of Plot: Petals of smaller size appear to be part of one category, Setosa. Petal width and length are alone enough to classify the type of flower even without the need of labels<b>

BoxPlot:

```python
sns.boxplot(data=scat)
plt.title("Boxplot of iri features (All)")
plt.xticks(rotation=45)
plt.title("Distribution of Iris features", fontsize=16, weight='bold')
plt.ylabel("Value range", fontsize=12)
plt.xlabel("Features", fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()
```

<b>Conclusion : Sepal Width shows outliers with are outside of the IQ range and may show that there is some ambiguity in the dataset </b>

```python
plt.figure(figsize=(10,6))

plt.subplot(2, 2, 1)
sns.histplot(scat["sepal length (cm)"], bins=20, kde=True, color="skyblue")
plt.title("Sepal Length")

plt.subplot(2, 2, 2)
sns.histplot(scat["sepal width (cm)"], bins=20, kde=True, color="lightgreen")
plt.title("Sepal Width")

plt.subplot(2, 2, 3)
sns.histplot(scat["petal length (cm)"], bins=20, kde=True, color="salmon")
plt.title("Petal Length")

plt.subplot(2, 2, 4)
sns.histplot(scat["petal width (cm)"], bins=20, kde=True, color="plum")
plt.title("Petal Width")

plt.tight_layout()
plt.show()
```

<b>Conclusion: Histogram for Petal lenght and width shows there are few missing values inside the dataset which can be considered either as outliers</b>
